```python
# 2.3.3 Plotting using Logarithmic Axes
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=12, label="First")
plt.plot(x, y2, "gs-", linewidth=2, markersize=12, label="Second")

plt.xlabel("$X$")  # LaTex
plt.ylabel("Y")    # showing in different font

plt.axis([-0.5, 10.5, -5, 105])  #[xmin, xmax, ymin, ymax]

plt.legend(loc="upper left")   # requries label attri in .plot()

plt.savefig("myplot.pdf") 
# save into working dir where launched Python
```

```python
x = np.logspace(-1, 1, 40)
y1 = x**2.0
y2 = x**1.5
plt.xlabel("X'")
plt.ylabel("Y'")
plt.loglog(x, y1, "bo-", linewidth=2, markersize=2, label="First")
plt.loglog(x, y2, "gs-", linewidth=2, markersize=2, label="Second")
```

```python
x = np.logspace(0, 1, 10)
y = x**2
plt.loglog(x, y, "bo-")
```

```python
# Histograms
x = np.random.normal(size=1000) 
#1000 numbers for normal distribution, mean=0 and variance=1
plt.hist(x); # y-axis is the number of observations that fall in each bin
```

```python
plt.hist(x, density=True); # y-axis is proportion of observations in that bin
```

```python
plt.hist(x, density=True, bins=np.linspace(-5,5,21)); #21 specifies 20 bins
```

```python
# Histogram example using gamma distribution
x = np.random.gamma(2, 3, 100000)
plt.hist(x, 30);
```

```python
plt.hist(x, 30, density=True, cumulative=True, histtype="step");
```

```python
# Generate 1 figure with 4 panels where each histogram in its' own subplot
plt.figure()
plt.subplot(221)
plt.hist(x, bins = 30);
plt.subplot(222)
plt.hist(x, bins = 30, density = True);
plt.subplot(223)
plt.hist(x, bins = 30, cumulative = 30);
plt.subplot(224)
plt.hist(x, bins = 30, density = True, cumulative = True, histtype = "step");
```
